# Parte 1. Procesamiento de datos
Para el procesamiento de datos primero se instalaron las librerias: 
- *bibtexparser* para leer y escribir archibos BibTex
- *pandas* para análisis y manipulación de datos
- *tqdm* añade barras de progreso en bucles o iterables
- *requests* permite hacer peticiones HTTP
Con el comando: 
    <pip install bibtexparser pandas tqdm requests>

```python
import bibtexparser
import pandas as pd

# Ruta del archivo
file_path = "../data_raw/bctz_library_raw.bib"

with open(file_path, encoding="utf-8") as bib_file:
    bib_database = bibtexparser.load(bib_file)

# Convertir a DataFrame
df = pd.DataFrame(bib_database.entries)

print("Número de registros:", len(df))
df.head()
```

```python
print("Abstracts faltantes:", df['abstract'].isna().sum())
print("DOI faltantes:", df['doi'].isna().sum())
print("Años únicos:", df['year'].nunique())
```
